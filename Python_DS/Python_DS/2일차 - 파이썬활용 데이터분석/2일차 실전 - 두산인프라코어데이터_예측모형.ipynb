## T검정:
- 공장에 따른 claim count의 값의 차이가 통계적으로 유의한지 알기 위해
- 먼저, groupby, agg함수를 사용하여 모델별 평균과 표준편차를 구하고, 
- 보통 표준편차가 작으면 합동분산 T검정을 실시함
- p_value - 0.05이하로 나와야 유의함 수준함 - > 그룹간 CTQ값의 차이가 통계적으로 유의함

```python
ff.groupby(['has_telemetry'])[['claim_count']].agg(['std', 'mean'])
```

```python
model_1 = ff.loc[ff['has_telemetry'] == True, 'claim_count']
model_2 = ff.loc[ff['has_telemetry'] == False, 'claim_count']
model_2
```

```python
import statsmodels.api as sm
```

```python
tstat, pvalue, diff = sm.stats.ttest_ind(model_1, model_2)
print('T값은 %.3f 이고 Pvalue는 %.5f' % (tstat, pvalue))
```
