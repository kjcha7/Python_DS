```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url ='https://search.shopping.naver.com/search/all.nhn?origQuery=%EC%84%B8%ED%83%81%EA%B8%B0&pagingIndex=1&pagingSize=40&viewType=list&sort=rel&frm=NVSHATC&query=%EC%84%B8%ED%83%81%EA%B8%B0'
```

```python
r = requests.get(url)
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
soup
```

```python
goods = soup.select('ul.goods_list li div.info a.tit')
goods
```

```python
print("++네이버 쇼핑 세탁기 순위++")
num=0
for i in goods:
    num +=1
    print(num, i['title'])
```
